```python
import pandas as pd

def compute_z_scores_and_correlations(df, columns):
    # Dictionary to store original and normalized ranges
    ranges = {'original': {}, 'normalized': {}}
    
    for column in columns:
        # Calculate Z-score normalization
        df[f'{column}_z'] = (df[column] - df[column].mean()) / df[column].std()

        # Store original range
        original_min = df[column].min()
        original_max = df[column].max()
        ranges['original'][column] = (original_min, original_max)

        # Store normalized range
        normalized_min = df[f'{column}_z'].min()
        normalized_max = df[f'{column}_z'].max()
        ranges['normalized'][column] = (normalized_min, normalized_max)

        # Print ranges for each column
        print(f'Original range for {column}: {original_min} to {original_max}')
        print(f'Normalized range for {column}_z: {normalized_min} to {normalized_max}')
    
    # Compute correlation matrix for normalized columns
    z_columns = [f'{col}_z' for col in columns]
    correlation_matrix = df[z_columns].corr()
    
    return df, correlation_matrix

    # Define metrics to be analyzed
metrics = ['nltk_opinion_lexicon_net', 'vader_neg', 
               'vader_neu', 'vader_pos', 'vader_compound', 'openai_mean_sentiment']

    # Compute z-scores and correlation matrix
df, correlation_matrix = compute_z_scores_and_correlations(data, metrics)
df, correlation_matrix
# Example usage
# Assuming df is already defined and columns list is available
# columns = ['nltk_opinion_lexicon_net', 'vader_neg', 'vader_neu', 'vader_pos', 'vader_compound', 'openai_mean_sentiment']
# df, correlation_matrix = compute_z_scores_and_correlations(df, columns)
```
